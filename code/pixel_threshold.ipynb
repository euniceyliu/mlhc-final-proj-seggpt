```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename), cv2.IMREAD_GRAYSCALE)
        if img is not None:
            images.append(img)
    return images
```

plot the histogram of the pixel density

```python
def plot_histograms(images, class_label):
    """
    Plot histograms for the pixel intensity distribution of a list of images.

    This function iterates through each image in the provided list, calculates the histogram of pixel intensities,
    and then plots these histograms. Each histogram represents the frequency of pixel intensities within the image,
    ranging from 0 to 255.

    Parameters:
        images (list of ndarray): A list of images where each image is represented as a numpy array.

    Returns:
        None: This function does not return any values but displays histograms of pixel intensities for each image.

    Each histogram is displayed in a separate figure with pixel intensities on the x-axis and their corresponding
    frequency on the y-axis.
    """
    for img in images:
        plt.figure()
        plt.hist(img.ravel(), bins=256, range=[0,256])
        plt.xlabel("Pixel Intensity")
        plt.ylabel("Frequency")
        plt.title(class_label)
        plt.show()
```

```python
# specific path to the class
folder_path = '/content/drive/MyDrive/MLHCImages/One-Shot_Gamma/Water'

# load images
images = load_images_from_folder(folder_path)

# plot
plot_histograms(images, "Water")

```
